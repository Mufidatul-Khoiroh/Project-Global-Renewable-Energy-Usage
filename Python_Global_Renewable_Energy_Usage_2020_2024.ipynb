# **Data Loading**
To display and examine the entire dataset.

```python
import pandas as pd

# Load the dataset
dataset = pd.read_csv('/content/Renewable_Energy_Usage_Sampled.csv')
display(dataset)
```

# **The Comparison of Sample Counts by Region**
To determine the sample counts across regions and identify any significant differences in sample counts and proportions between regions.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Group by 'Region' and count occurrences
region_counts = dataset['Region'].value_counts()

# Create a pie chart
plt.figure(figsize=(5, 5))
plt.pie(region_counts, labels=region_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('The Comparison of Sample Counts by Region')
plt.show()
```

# **Annual Sample Counts**
To display the comparison of sample counts across years.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sort by year for proper time-series visualization
year_counts = dataset['Year'].value_counts().sort_index()

# Create a bar chart
plt.figure(figsize=(8, 5))
year_counts.plot(kind='bar', color='green')
plt.title('Annual Sample Counts', loc='center', pad=30, fontsize=15, color='darkblue')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Samples', fontsize=12)
plt.ylim(ymin=0)
plt.xticks(rotation=0)
plt.show()
```

# **Total Annual Usage of Renewable Energy in Households Across Different Years**

To examine the annual pattern of household renewable energy consumption.

```python
# Group by 'Year' and sum 'Monthly_Usage_kWh'
total_usage_in_year = dataset.groupby('Year')['Monthly_Usage_kWh'].sum()

# Year
dataset['Year'] = dataset['Year'].astype(int)
year_counts = dataset['Year'].value_counts().sort_index()

# Create a line chart
plt.figure(figsize=(10, 6))
total_usage_in_year.plot(color='green', marker='o', linestyle='-', linewidth=2)
plt.title('Total Annual Usage of Renewable Energy in Households Across Different Years', color='blue', pad=20, fontsize=15)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Total Usage (kWh)', fontsize=12)
plt.grid(True)
plt.xticks(rotation=0)
plt.xticks(year_counts.index)
plt.show()
```

# **Total Household Usage of Renewable Energy in 2024**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Filter data for the year 2024
dataset_2024 = dataset[dataset['Year'] == 2024]

# Group by 'Energy_Source' and sum 'Monthly_Usage_kWh'
total_usage_by_source = dataset_2024.groupby('Energy_Source')['Monthly_Usage_kWh'].sum()

# Create a bar chart
plt.figure(figsize=(8, 5))
total_usage_by_source.plot(kind='bar', color='green')
plt.title('Total Household Usage of Renewable Energy in 2024', loc='center', pad=20, fontsize=15, color='darkblue')
plt.xlabel('Energy Source', fontsize=12)
plt.ylabel('Total Usage (kWh)', fontsize=12)
plt.ylim(ymin=0)
plt.xticks(rotation=0)
plt.show()
```
